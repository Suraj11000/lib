# Fill Between:

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]

y = [1,2,3,4,5,6]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=3,y2=5,color="y",)

plt.title("Fill Between")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5,6])

y = np.array([1,2,3,4,5,6])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="y",where=(x>=2) & (x<=4))

plt.title("Fill Between")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
