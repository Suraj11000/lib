# Stem:

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]

y = [2,4,1,4,2,3]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="y",label="Stem Plot",use_line_collection=False)

plt.legend()

plt.show()
```

```python
plt.stem(x,x,use_line_collection=False,orientation="horizontal")
```

```python
plt.stem(x,y,use_line_collection=False)
plt.show()
```
