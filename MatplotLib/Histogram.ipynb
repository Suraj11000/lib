# histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random

x = np.random.randint(10,40,(30))

print(x)
```

```python
#Bins = 1
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Bins = AUTO (0,100)
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,"auto", (0,100), edgecolor="black")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Cumulative :
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",edgecolor="black",cumulative=-1)

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Botttom
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange", edgecolor="black",bottom = 10)

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#align:
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black",align="right")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

# Histtype

```python
#Step
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black",histtype="step")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Barstacked
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black",histtype="barstacked")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Orientation:
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black",orientation = "horizontal")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Rwidth:
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black", rwidth = 0.8)

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
#Log :
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black", log = True, rwidth=0.8)

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.show()
```

```python
no = [28, 26, 31, 28, 27, 17, 38, 30, 22, 24, 16, 29, 35, 37, 14, 38, 36, 39, 37, 35, 12, 27, 19, 34, 38, 35, 29, 13, 27, 15, 14]
l=[10,20,30,40]
plt.hist(no,color="orange",bins=l, edgecolor="black",label="Suraj")

plt.title("Histogram")
plt.xlabel("NO")
plt.ylabel("Values")
plt.axvline(22,color="g", label = "axvline")
plt.legend()
plt.grid()
plt.show()
```
