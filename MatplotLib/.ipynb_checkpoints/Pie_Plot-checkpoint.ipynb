```python
import matplotlib.pyplot as plt

x = [10,20,30,40]

y = ["Suraj","Dhiraj","Sara","Riya"]

ex=[0.4,0.0,0.0,0.0]

c = ["r","g","b","y"]

plt.pie(x, labels = y,labeldistance = 0.8,explode=ex,colors = c,autopct="%0.0f%%",shadow = True,radius=1.2,startangle=90,textprops={"fontsize":15},counterclock=False,
       wedgeprops={"linewidth":5,"edgecolor":"orange"},rotatelabels = True)
plt.title("Pie Chart")
plt.legend(loc=2)
plt.show()
```
