# BOX PLOT :

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=["Suraj"],patch_artist=True,showmeans=True,sym="g*",
           boxprops=dict(color="r"), capprops=dict(color="g"),whiskerprops=dict(color="y"),
           )
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.5)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=["Suraj"],showmeans=True,whis=2)
plt.show()
```
