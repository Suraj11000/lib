# Stack

```python
import matplotlib.pyplot as plt

x=[0,20,30,40,50]
y=[5,30,10,50,15]
z=[2,10,8,50,25]
s=[35,20,50,30,15]

c=["r","b","y"]

l=["Suraj","Dhiraj","Sara","Riya"]
plt.stackplot(x,y,z,s,labels=l,colors=c)
plt.title("Stackplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.grid()
plt.show()
```
