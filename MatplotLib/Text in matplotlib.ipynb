# Title, xlabel, ylabel, text, Annoate

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[4,2,6,3,6]

plt.plot(x,y)

plt.title("Text in Graph",fontsize=20)
plt.xlabel("x-Axis",fontsize=15)
plt.ylabel("y_Axis",fontsize=15)

plt.text(3,6,"Max",fontsize=15,color="r",style="italic",bbox={"facecolor":"yellow"})

plt.annotate("Min",xy=(2,2),xytext=(3,3),arrowprops=dict(facecolor="green", shrink=100))

plt.legend(["data"],loc=2,facecolor="pink",edgecolor="black",shadow=True,framealpha=0.8)
plt.show()
```
