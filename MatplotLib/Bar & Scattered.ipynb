# Bar Plot

```python
import matplotlib.pyplot as plt
x = ["Suraj", "Dhiraj","Raj","Sara"]
y = [75,68,87,66]
c = ["r","g","y","b"]

plt.xlabel("Names", fontsize = 15)
plt.ylabel("Marks", fontsize = 15)
plt.title("Students", fontsize = 25)

plt.bar(x,y,color=c,width=0.5,align="edge",edgecolor="black",linewidth=2)

plt.show()
```

```python
plt.bar(x,y,color=c,edgecolor="black",linestyle=":",alpha=0.4,label="data")
plt.legend()
```

```python
plt.bar(x,y,color=c,edgecolor="black",label="data")
plt.legend()
plt.show()
```

```python
z=[44,56,23,76]

plt.bar(x,y,color="r",edgecolor="black",label="data")
plt.bar(x,z,color="y",edgecolor="r",label="data1")
plt.legend()
```

```python
import numpy as np
arr = np.arange(len(x))
 
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Student")
width = 0.2
arr1=[j+width for j in arr]

plt.bar(arr,y,width,color="r",label="d1")
plt.bar(arr1,z,width,color="y",label="d2")
plt.xticks(arr+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Student")
width = 0.2

plt.barh(x,y,width,color="r",label="d1")
plt.barh(x,z,width,color="y",label="d2")
plt.xticks(arr+width/2,x,rotation=10)
plt.legend()
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
x = [5,6,7,8]
y= [2,4,6,8]
color = ["r","g","b","y"]
size = [100,80,20,60]
plt.scatter(x,y,c=color,s=size)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```

```python
x = [5,6,7,8]
y= [2,4,6,8]
color = ["r","g","b","y"]
size = [100,80,20,60]
plt.scatter(x,y,c=color,s=size,alpha=0.4,marker="+")
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```

```python
x = [5,6,7,8]
y= [2,4,6,8]
color = ["r","g","b","y"]
size = [300,200,250,280]
plt.scatter(x,y,c=color,s=size,marker="*",edgecolor="orange",linewidth=2)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```

```python
x = [5,6,7,8]
y= [2,4,6,8]
color = [22,11,33,88]
size = [300,200,250,280]
plt.scatter(x,y,c=color,s=size,cmap="BrBG")
plt.colorbar()
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```

```python
x = [5,6,7,8]
y= [2,4,6,8]
color = [22,11,33,88]
size = [300,200,250,280]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```

```python
x = [5,6,7,8]
y= [2,4,6,8]
color = [22,11,33,88]
size = [300,200,250,280]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
cb = plt.colorbar()
cb.set_label("ColorBar", fontsize=15)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```

```python
x = [5,6,7,8]
y= [2,4,6,8]
z=[1,3,5,7]
color = [22,11,33,88]
size = [300,200,250,280]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
plt.scatter(x,z,color="r",s=size,alpha=0.2)
cb = plt.colorbar()
cb.set_label("ColorBar", fontsize=15)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Months")
plt.show()
```
