# Axis:

1) Ticks

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,1,4,5]

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4]
y=[2,1,4,5]

plt.plot(x,y)

l = ["Suraj","Dhiraj","sara","Riya"]

plt.xticks(x,labels=l)
plt.yticks(x)

plt.show()
```

# Lim:

```python
x=[1,2,3,4,5]
y=[2,1,4,5,3]

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,10)

plt.show()
```

# Axis:

```python
x=[1,2,3,4,5]
y=[2,1,4,5,3]

plt.plot(x,y)

plt.axis([0,10,0,7])

plt.show()
```
