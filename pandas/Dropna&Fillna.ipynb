# Dropna

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\shind\\Desktop\\pandas\\max.csv")
df
```

```python
# this will frop row by default
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how="any")
```

```python
df.dropna(how="all")
```

```python
df.dropna(subset=["Sex"])
```

```python
df.dropna(inplace=True)
df
```

```python
df.dropna(thresh=1)
```
