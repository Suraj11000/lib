# Write CSV

```python
import pandas as pd

dict={
    "A":[1,4,7,8,5],
    "B":[9,8,7,5,2],
    "C":[5,7,3,8,4]
}
df = pd.DataFrame(dict)
print(df)
print()
df.to_csv("Data.csv")
```

```python
df.to_csv("data1.csv",index=False)
df
```

```python
df.to_csv("data2.csv",index=False,header=["Suraj","Djiraj","Sara"])
df
```
