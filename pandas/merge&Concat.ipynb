# Merge:

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[6,7,8,9]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[16,47,98,49]})

pd.merge(df1,df2,how="inner")
```

```python
pd.merge(df1,df2,how="left")
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,how="outer")
```

```python
pd.merge(df1,df2,how="outer", indicator=True)
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[6,7,8,9]})
df2 = pd.DataFrame({"A":[1,2,3,5],"B":[16,47,98,49]})

pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=("name","ID"))
```

# Concat

```python
df1 = pd.Series([1,2,3,4])
df2 = pd.Series([11,45,6,55])

pd.concat([df1,df2])
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[6,7,8,9]})
df2 = pd.DataFrame({"A":[1,2,3,5],"B":[16,47,98,49]})

pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[6,7,8,9]})
df2 = pd.DataFrame({"A":[1,2],"C":[16,47]})

pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1,join="inner")
```

```python
pd.concat([df1,df2],axis=1,keys=["d1","d2"])
```
