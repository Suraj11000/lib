# Replace:

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\shind\\Desktop\\pandas\\max.csv")
df
```

```python
df.replace(to_replace="Shelby",value="Suraj")
```

```python
df.replace([1,2,3],51)
```

```python
df.replace("[A-Za-z]","Suraj",regex=True)
```

```python
df.replace({"Sex":"[A-Z]"},51,regex=True)
```

```python
df.replace("Male",method="ffill")
```

```python
df.replace("Male",method="bfill")
```

```python
df.replace(1,method="ffill",limit=2)
```

```python
df.replace(1,method="ffill",limit=3,inplace=True)
df
```
