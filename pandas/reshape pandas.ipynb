# Melt

```python
import pandas as pd
df = pd.DataFrame({"Day":[1,2,3,4,5,6,7],
                  "Eng":[19,12,13,16,14,11,9],
                  "maths":[20,13,16,15,18,17,19]})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=["Eng"])
```

```python
pd.melt(df,id_vars=["Day"])
```

```python
pd.melt(df,id_vars=["Day"],var_name="Subjects")
```

```python
pd.melt(df,id_vars=["Day"],var_name="Subjects",value_name="Marks")
```

# Privit()

```python
df = pd.DataFrame({"Day":[1,2,3,4,5,6,7],
                   "st_name":["a","b","c","a","b","c","a"],
                  "Eng":[19,12,13,16,14,11,9],
                  "maths":[20,13,16,15,18,17,19]})
df
```

```python
df.pivot(index="Day",columns="st_name")
```

```python
df.pivot(index="Day",columns="st_name",values="Eng")
```

```python
df.pivot_table(index="st_name", columns="Day",aggfunc="mean")
```

```python
df.pivot_table(index="st_name", columns="Day",aggfunc="sum",)
```

```python
df.pivot_table(index="st_name", columns="Day",aggfunc="mean",margins="True")
```
